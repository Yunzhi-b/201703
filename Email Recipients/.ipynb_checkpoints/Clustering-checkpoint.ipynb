# Combine dataframes

```python
def transform(dataframe):
    df = dataframe.copy()
    training_arrays = []
    for index, series in df.iterrows():
        row = series.tolist()
        sender = row[0]
        ids = row[1:][0].split(' ')
        for id in ids:
            training_arrays.append([sender,int(id)])
    return pd.DataFrame(training_arrays,columns=['sender','mid'])
```

### Given wordlist(filt), compute matrix TF-IDF by SENDER, RECIPIENT


```python
# for all messages m_u sent by user u, we derived the message’s TF-IDF vector representation from its textual contents 
# and then normalized the vector to length 1. 

# We then built, for each recipient r_i in the Address Book AB(u), a TF-IDF centroid vector C(r_i), by summing up the normalized 
# TF-IDF vectors of all messages that were sent from u to r_i.

# When testing, for each test message we computed the cosine similarities between the message’s TF-IDF vector representation and the
# |AB(u)| Centroid vectors. The |AB(u)| recipients are then ranked according to these cosine similarity scores
```

### create a dictionary of wordlist
wordlist["s1"] stores all the key words of mails sent by s1

```python
from nltk.probability import FreqDist

def wordlist_bySender(sender,hf=None,lf=None):
    
    idx = index_senders[sender]

    words = [word for ix in idx for word in filt_tr_subwords[ix]]
    total_words = len(words)

    words_unique = set(words)
    
    fdist1 = FreqDist(words)
       #length>15
    long_words = [w for w in words_unique if len(w) > 15]
    len(sorted(long_words))

    #length<2
    short_words = [w for w in words_unique if len(w) < 4]
    len(sorted(short_words))

    #filter

    filt = set(words_unique)-set(long_words)-set(short_words) 
    
    #remove high frequency and low frequency
    if hf: 
        w_hf, _ = zip(*fdist1.most_common(hf))
        filt = filt-set(w_hf)
    if lf:
        w_lf = list(filter(lambda x: fdist1[x]<=lf, fdist1))
        filt = filt-set(w_lf)

    #remove numbers
    filt = [w for w in filt if w.isalpha()]
    
    return filt
```

```python
wordlist = {}

for s in senders:
    wordlist[s] = wordlist_bySender(s,lf=3)
    
len_wordlist = [len(wordlist[s]) for s in senders]
print (min(len_wordlist),max(len_wordlist))
```

### Compute Centroids of (sender, recipient)

```python
from numpy.linalg import norm
"""
objectif: for all messages m_u sent by user u, we derived the message’s TF-IDF vector representation from its textual contents 
and then normalized the vector to length 1. 

"""

# return a dictonary of words in form of {word: number of documents containing this word + 1}
def IDF_bySender(sender,X,words):
    IDFs={}
    idx = index_senders[sender]
    X = [X[ix] for ix in idx]
    n=float(len(X))
    
    for word in words:
        IDFs.setdefault(word, 1)
    
    #print "sender", sender
    # print "total lines %.0f"%n 
    for idx,line in enumerate(X):
        #if idx % 100==0:
        #    print "precessing...%d" %idx
            
        for word in words:
            if word in line:
                IDFs[word]=IDFs[word]+1
    
    for word in words:
        IDFs[word]=np.log(n/IDFs[word])    
    return IDFs

# return a line of Tf-idf (normalized). It's of shape (1,m) with m the number of words we use.
def TFIDF_oneLine(line,words,IDFs):
    length=float(len(line))+1.
    tfidfs = [line.count(word)/length*IDFs[word] for word in words]
    norm_tfidfs = norm(tfidfs)
    if norm_tfidfs==0:
        return tfidfs
    else:
        return tfidfs/norm_tfidfs
    

# return a matrix of Tf-idfs of shape (n,m). with n number of sample, m the number of words we use.
def matrixOfTfidf_bySender(sender,X,words):
    idfs=IDF_bySender(sender,X,words)

    idx = index_senders[sender]
        
    X = [X[ix] for ix in idx]
    result = [TFIDF_oneLine(line,words,idfs) for line in X]
    return result
```

```python
tfidf_tr_bySender ={}

print "total senders %d" %len(senders)

for i, s in enumerate(senders):
    if i % 10==0:
        print "precessing...%d" %i
        
    tfidf_tr_bySender[s] = matrixOfTfidf_bySender(s, filt_tr_subwords, wordlist[s])   
```

### Transform test file to tfidf matrix

```python
def matrixOfTfidf_bySender_test(sender,X_train,X_test,words):
    idfs=IDF_bySender(sender,X_train,words)

    idx = index_senders_test[sender]
        
    X_test = [X_test[ix] for ix in idx]
    result = [TFIDF_oneLine(line,words,idfs) for line in X_test]
    return result
```

```python
tfidf_test_bySender={}

print "total senders in test", len(senders_test)


for i, s in enumerate(senders_test):
    if i % 10==0:
        print "precessing...%d" %i

    tfidf_test_bySender[s] = matrixOfTfidf_bySender_test(s, filt_tr_subwords , filt_test_subwords, wordlist[s])
```

### Prediction 

- centroids([sender][recipient]), representation)
- tfidf_test_bySender(sender): representation of test set
- index_senders_test(sender): index of sender in test set
- senders_test: list of unique senders in test set

```python
from scipy import spatial

dict_test_recps={}

for s in senders_test:
    
    if s:#!='mike.maggi@enron.com':
        
        X_value = centroids[s].values()  # centroids where key (sender,_)=s
        X_key = centroids[s].keys()
        Y = tfidf_test_bySender[s]
        D=[]
    
        for i in range(len(Y)):
            tmp = [1- spatial.distance.cosine(Y[i], v) for v in X_value]    # cos(vect,matrix)
            
            j = np.argsort(tmp)[::-1][:10]
            
            recps = [X_key[k] for k in j ]
            D.append(recps)
    
        dict_test_recps[s] = D
```

```python
output={}

for s in senders_test:
    if s:#!='mike.maggi@enron.com':
        n_mail_bySender = len(dict_test_recps[s])
    
        for i in range(n_mail_bySender):
            output[index_senders_test[s][i]] = dict_test_recps[s][i]
```
