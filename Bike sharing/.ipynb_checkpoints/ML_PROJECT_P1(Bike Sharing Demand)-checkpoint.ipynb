```python
options_weekdays = {'Monday' : 1,
            'Tuesday' : 2,
            'Wednesday' : 3,
            'Thursday' : 4,
            'Friday' : 5,
            'Saturday' : 6,
            'Sunday' : 7,
}

import datetime

def gettime(dataframe):
    df = dataframe.copy()
    df['year'] = pd.DatetimeIndex(df['datetime']).year
    df['month'] = pd.DatetimeIndex(df['datetime']).month
    df['day'] = pd.DatetimeIndex(df['datetime']).day
    n = len(df)
    for i in np.arange(n):
        tmp = df.ix[i].datetime
        hour = int(datetime.datetime.strptime(tmp, '%Y-%m-%d %H:%M:%S').strftime('%H'))
        weekday = datetime.datetime.strptime(tmp, '%Y-%m-%d %H:%M:%S').strftime('%A')
        df.set_value(i, 'hour', hour)
        df.set_value(i, 'weekday', options_weekdays[weekday])
    return df
```

## Random Forest 

### 1. Evaluation metric RMSLE

```python
#A function to calculate Root Mean Squared Logarithmic Error (RMSLE)
def rmsle(y, yhat):
    y = y.ravel()
    yhat = yhat.ravel()
    assert len(y) == len(yhat)
    return np.sqrt(np.mean(np.power(np.log1p(y)-np.log1p(yhat), 2)))
```

```python
from sklearn.metrics import make_scorer
rmsle_loss = make_scorer(rmsle, greater_is_better=False)
```

**mse**

```python
from sklearn.metrics import mean_squared_error
mse_err = make_scorer(mean_squared_error,greater_is_better=False)
```

- **Check Stationarity of a Time Series**

A TS is said to be stationary if its statistical properties such as mean, variance remain constant over time. Most of the TS models work on the assumption that the TS is stationary. 
For practical purposes we can assume the series to be stationary if it has constant statistical properties over time. 

i.e. the following:
- 1.constant mean 
- 2.constant variance
- 3.an autocovariance that does not depend on time.

```python
fig = plt.figure(figsize=(15,4))
plt.subplot(131)
plt.plot(df_train_ts["count"])
plt.title('number of total rentals')
plt.subplot(132)
plt.plot(df_train_ts["registered"])
plt.title('number of registered user rentals initiated')
plt.subplot(133)
plt.plot(df_train_ts["casual"])
plt.title('number of non-registered user rentals initiated')
```

More formally, we can check stationarity using the following:

- **Plotting Rolling Statistics**: We can plot the moving average or moving variance and see if it varies with time. At any instant t, well take the average/variance of the last year, i.e. last **1** months. But again this is more of a visual technique.
- **Dickey-Fuller Test**: This is one of the statistical tests for checking stationarity. Here the null hypothesis is that the TS is non-stationary. The test results comprise of a Test Statistic and some Critical Values for difference confidence levels. If the Test Statistic is less than the Critical Value, we can reject the null hypothesis and say that the series is stationary.

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, window=24):
    
    #Determing rolling statistics
    # 1 month = 24*18 =432 records
    rolmean = pd.rolling_mean(timeseries, window)
    rolstd = pd.rolling_std(timeseries, window)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation 2011-01')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(df_train_ts['count'],432)
```

```python
test_stationarity(df_train_ts["2011-01"]['count'],24)
```
