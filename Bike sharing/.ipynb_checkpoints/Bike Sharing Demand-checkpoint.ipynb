**Description**

You are provided hourly rental data spanning two years. For this competition, the training set is comprised of the first 19 days of each month, while the test set is the 20th to the end of the month.

**Data Fields**

- **datetime** - hourly date + timestamp  
- **season** -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
- **holiday** - whether the day is considered a holiday
- **workingday** - whether the day is neither a weekend nor holiday
- **weather** - 1: Clear, Few clouds, Partly cloudy, Partly cloudy 
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist 
3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds 
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
- **temp** - temperature in Celsius
- **atemp** - "feels like" temperature in Celsius
- **humidity** - relative humidity
- **windspeed** - wind speed
- **casual** - number of non-registered user rentals initiated
- **registered** - number of registered user rentals initiated
- **count** - number of total rentals

**Objective**

You must predict the total **count** of bikes rented during each hour covered by the test set

### 2.2 Generate new variables

```python
options_weekdays = {'Monday' : 1,
            'Tuesday' : 2,
            'Wednesday' : 3,
            'Thursday' : 4,
            'Friday' : 5,
            'Saturday' : 6,
            'Sunday' : 7,
}


import datetime

def gettime(dataframe):
    df = dataframe.copy()
    df['year'] = pd.DatetimeIndex(df['datetime']).year
    df['month'] = pd.DatetimeIndex(df['datetime']).month
    df['day'] = pd.DatetimeIndex(df['datetime']).day
    n = len(df)
    for i in np.arange(n):
        tmp = df.ix[i].datetime
        hour = int(datetime.datetime.strptime(tmp, '%Y-%m-%d %H:%M:%S').strftime('%H'))
        weekday = datetime.datetime.strptime(tmp, '%Y-%m-%d %H:%M:%S').strftime('%A')
        df.set_value(i, 'hour', hour)
        df.set_value(i, 'weekday', options_weekdays[weekday])
    return df

# 07:00:00 - 22:00:00 Jours 
```

```python
from pandas.core.frame import DataFrame

def addMeanValues(df_train, df_test, bySeason=True, byHoliday=True, byWorkingday=True, byWeather=True, byDay=True, byMonth=True, byYear=True, byHour=True,byWeekday=True):
    df = df_train.copy()
    df_2 = df_test.copy()
    if bySeason:
        df_meanBySeason = DataFrame({'meanBySeason':df.groupby(['season'])['count'].mean()}).reset_index()
        df = df.join(df_meanBySeason.set_index('season'),on='season')
        df_2 = df_2.join(df_meanBySeason.set_index('season'),on='season')
    if byHoliday:
        df_meanByHoliday = DataFrame({'meanByHoliday':df.groupby(['holiday'])['count'].mean()}).reset_index()
        df = df.join(df_meanByHoliday.set_index('holiday'),on='holiday')
        df_2 = df_2.join(df_meanByHoliday.set_index('holiday'),on='holiday')
    if byWorkingday:
        df_meanByWorkingday = DataFrame({'meanByWorkingday':df.groupby(['workingday'])['count'].mean()}).reset_index()
        df = df.join(df_meanByWorkingday.set_index('workingday'),on='workingday')
        df_2 = df_2.join(df_meanByWorkingday.set_index('workingday'),on='workingday')
    if byWeather:
        df_meanByWeather = DataFrame({'meanByWeather':df.groupby(['weather'])['count'].mean()}).reset_index()
        df = df.join(df_meanByWeather.set_index('weather'),on='weather')
        df_2 = df_2.join(df_meanByWeather.set_index('weather'),on='weather')
    if byDay:
        df_meanByDay = DataFrame({'meanByDay':df.groupby(['day'])['count'].mean()}).reset_index()
        df = df.join(df_meanByDay.set_index('day'),on='day')
        df_2 = df_2.join(df_meanByDay.set_index('day'),on='day')
    if byMonth:
        df_meanByMONTH = DataFrame({'meanByMonth':df.groupby(['month'])['count'].mean()}).reset_index()
        df = df.join(df_meanByMONTH.set_index('month'),on='month')
        df_2 = df_2.join(df_meanByMONTH.set_index('month'),on='month')
    if byYear:
        df_meanByYear = DataFrame({'meanByYear':df.groupby(['year'])['count'].mean()}).reset_index()
        df = df.join(df_meanByYear.set_index('year'),on='year')
        df_2 = df_2.join(df_meanByYear.set_index('year'),on='year')     
    if byHour:
        df_meanByHour = DataFrame({'meanByHour':df.groupby(['hour'])['count'].mean()}).reset_index()
        df = df.join(df_meanByHour.set_index('hour'),on='hour')
        df_2 = df_2.join(df_meanByHour.set_index('hour'),on='hour')
    if byWeekday:
        df_meanByWeekday = DataFrame({'meanByWeekday':df.groupby(['weekday'])['count'].mean()}).reset_index()
        df = df.join(df_meanByWeekday.set_index('weekday'),on='weekday')
        df_2 = df_2.join(df_meanByWeekday.set_index('weekday'),on='weekday')
    return df,df_2
```

```python
df_train_v1 = gettime(df_train)
df_test_v1 = gettime(df_test)
```

## 2 Training

### 2.1 Evaluation metric RMSLE

```python
#A function to calculate Root Mean Squared Logarithmic Error (RMSLE)
def rmsle(y, yhat):
    y = y.ravel()
    yhat = yhat.ravel()
    assert len(y) == len(yhat)
    return np.sqrt(np.mean(np.power(np.log1p(y)-np.log1p(yhat), 2)))
```

```python
from sklearn.metrics import make_scorer
rmsle_loss = make_scorer(rmsle, greater_is_better=False)
```

### 2.3 RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA

Xtrain, Xvalid,ytrain,yvalid = train_test_split(X_train,y_train, test_size=0.2, random_state=10)
Xtest = X_test.copy()
```

** With PCA ** 


```python
param = {
    'pca__n_components': [15,11],
    #'rf__n_estimators' : range(100,140,10),
    'rf__max_features': [i/10.0 for i in range(5,7)],
    
}
```

```python
estimator = Pipeline([
            ('pca', PCA()),
            ('rf', RandomForestRegressor(n_estimators=130, criterion='mse'))])
gsearch = GridSearchCV(estimator, param, scoring = rmsle_loss, cv = 5)
gsearch.fit(Xtrain,ytrain)
gsearch.best_params_
```

### 2.4 XGBoost

```python
def rmsleErr(preds, dtrain):
    labels = dtrain.get_label()
    assert len(preds) == len(labels)
    return 'error', np.sqrt(np.mean(np.power(np.log1p(preds)-np.log1p(labels), 2)))
```

```python
Xtrain, Xvalid,ytrain,yvalid = train_test_split(X_train,y_train, test_size=0.2, random_state=42)
Xtest = X_test.copy()
```

```python
param = { "eta": 0.05,   #learning_rate
          "max_depth": 8,
          "subsample": 0.9,
          "lambda": 0.1  #L2 regularization term on weights
          }
num_round=500
```

```python
import xgboost as xgb

dtrain = xgb.DMatrix(Xtrain,label=ytrain)
dvalid = xgb.DMatrix(Xvalid, label=yvalid)
watchlist = [(dtrain, 'train'), (dvalid, 'eval')]

# prediction data
dtest = xgb.DMatrix(Xtest)
```

```python
bst = xgb.train(param, dtrain, num_round,evals=watchlist, early_stopping_rounds=50,feval=rmsleErr,verbose_eval=True)
```

```python
yhat = bst.predict(dvalid,ntree_limit=bst.best_iteration)
yhat[yhat<0]=0
error = rmsle(yvalid, yhat)
print(error)
```

```python
ypred = bst.predict(dtest,ntree_limit=bst.best_iteration)
ypred[ypred<0]=0
```
